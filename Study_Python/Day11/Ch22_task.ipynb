여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
def Sum(*args):
    sum = 0
    for arg in args:
        sum += arg
    return sum


print("더하고 싶은 수의 개수 입력 : ", end = "")
numList = []
n = int(input())
for i in range(n):
    print("숫자 입력 : ", end = "")
    numList.append(int(input()))
print(Sum(*numList))
```
