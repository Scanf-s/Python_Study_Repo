```python
import mymath

# 삼각형의 넓이 계산
tri_base = 10
tri_height = 5
triangleArea = mymath.triangle_area(tri_base, tri_height)
print(f"삼각형의 넓이: {triangleArea}")

# 원의 넓이 계산
radius = 7
circleArea = mymath.circle_area(radius)
print(f"원의 넓이: {circleArea}")

# 직육면체의 넓이 계산
length = 4
width = 5
height = 6
cuboidArea = mymath.cuboid_area(length, width, height)
print(f"직육면체의 넓이: {cuboidArea}")
```
